### Exercício 1.3 - Calculadora de IMC

Crie um programa que calcule o Índice de Massa Corporal (IMC) de uma pessoa.

A fórmula do IMC é: **IMC = peso / (altura)²**

O programa deve:
1. Solicitar o peso em quilogramas (float)
2. Solicitar a altura em metros (float)
3. Calcular o IMC
4. Exibir o resultado formatado com 2 casas decimais

**Exemplo de entrada**:
```
Digite seu peso em kg: 70.5
Digite sua altura em metros: 1.75
```

**Saída esperada**:
```

```

- Use `**` para calcular potência (altura ao quadrado)

### Exercício 2.1 - Transformações de Texto

Dada a string: `"  python para dados  "`

Realize as seguintes transformações e exiba cada resultado:
1. Remova os espaços nas extremidades
2. Converta para maiúsculas
3. Converta para minúsculas
4. Capitalize (primeira letra maiúscula)

**Saída esperada**:

Crie um relatório de vendas formatado usando f-strings com os seguintes dados:

- Nome do produto: "Notebook Gamer"
- Quantidade vendida: 125
- Preço unitário: 3299.99
- Desconto aplicado: 0.15 (15%)
- Total bruto: quantidade × preço
- Total com desconto: total bruto × (1 - desconto)

Exiba um relatório formatado com:
1. Nome do produto centralizado (20 caracteres)
2. Quantidade com preenchimento de zeros (3 dígitos)
3. Preço unitário como moeda (R$ com 2 decimais)
4. Desconto como porcentagem (2 decimais)
5. Total bruto com separador de milhar e 2 decimais
6. Total com desconto com separador de milhar e 2 decimais

**Saída esperada**:
```
========== RELATÓRIO DE VENDAS ==========
Produto:    Notebook Gamer    
Quantidade: 125 unidades
Preço Unit.: R$ 3299.99
Desconto:   15.00%
Total Bruto: R$ 412,498.75
Total Final: R$ 350,623.94
=========================================
```

**Dicas**:
- Use f-strings com formatação: `{valor:.2f}`, `{valor:,}`, `{texto:^20}`
- Calcule os totais antes de formatar
- Use alinhamento centralizado para o nome do produto

### Exercício 3.4 - Calculadora de Desconto Progressivo

Crie uma calculadora de desconto que aplica descontos progressivos baseados no valor da compra:

- Compras até R$ 100: sem desconto
- Compras entre R$ 100 e R$ 500: 5% de desconto
- Compras entre R$ 500 e R$ 1000: 10% de desconto
- Compras acima de R$ 1000: 15% de desconto

Além disso, se o cliente for VIP (True/False), adicione 2% de desconto extra.

O programa deve:
1. Solicitar o valor da compra
2. Solicitar se é cliente VIP
3. Calcular o desconto baseado na faixa de valor
4. Adicionar desconto VIP se aplicável
5. Calcular o valor final
6. Exibir: valor original, desconto aplicado (%), valor do desconto (R$), valor final

**Exemplo de entrada**:
```
Digite o valor da compra: R$ 750.00
É cliente VIP? (True/False): True
```

**Saída esperada**:
```
Valor original: R$ 750.00
Desconto aplicado: 12.00%
Valor do desconto: R$ 90.00
Valor final: R$ 660.00
```

**Dicas**:
- Use operadores relacionais para verificar faixas (`>`, `<=`, `and`)
- Use operadores aritméticos para cálculos
- Combine múltiplas condições com operadores lógicos
- Lembre-se da precedência: parênteses > aritméticos > relacionais > lógicos

````python
valor_compra = float(input("Digite o valor da compra: R$ "))
e_vip_input = input("É cliente VIP? (True/False): ")
e_vip = e_vip_input == "True"

# Determinando o desconto baseado na faixa de valor
desconto = 0.0  # Inicializando desconto

if valor_compra <= 100:
    desconto = 0.0  # Sem desconto
elif valor_compra <= 500:
    desconto = 0.05  # 5% de desconto
elif valor_compra <= 1000:
    desconto = 0.10  # 10% de desconto
else:
    desconto = 0.15  # 15% de desconto

# Adicionando desconto VIP se aplicável
if e_vip:
    desconto += 0.02  # Adiciona 2% de desconto extra

# Calculando valores
valor_desconto = valor_compra * desconto
valor_final = valor_compra - valor_desconto

# Exibindo o resultado
print(f"Valor original: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto:.2%}")
print(f"Valor do desconto: R$ {valor_desconto:.2f}")
print(f"Valor final: R$ {valor_final:.2f}")

````
